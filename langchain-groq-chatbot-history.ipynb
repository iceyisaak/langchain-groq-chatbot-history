## Message History

```python
from langchain_community.chat_message_histories import ChatMessageHistory
from langchain_core.chat_history import BaseChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory
```

```python
store={}

def get_session_history(session_id:str)->BaseChatMessageHistory:
    if session_id not in store:
        store[session_id]=ChatMessageHistory()
    return store[session_id]
```

```python
config={'configurable':{'session_id':'chat1'}}
```

```python
with_message_history=RunnableWithMessageHistory(llm,get_session_history)
```

```python
response = with_message_history.invoke([
    HumanMessage(
        content="Hi, I'm Isaak and I'm and AI Engineer."
        ),
    ],
    config=config
)
```

```python
response
```

```python
response.content
```

```python
with_message_history.invoke([
    HumanMessage(
        content="What's my name?"
        ),
    ],
    config=config
)
```

## Prompt Template

```python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
```

```python
prompt=ChatPromptTemplate.from_messages([
    ("system","You are a helpful assistant. Answer all the questions to the best of your ability."),
    MessagesPlaceholder(variable_name='messages')
])
```

```python
chain=prompt|llm
```

```python
chain.invoke({
    "messages":[
        HumanMessage(
            content="Hi, I'm Isaak!"
        )
    ]
})
```

```python
with_message_history=RunnableWithMessageHistory(chain,get_session_history)
```

```python
config_3={'configurable':{'session_id':'chat3'}}
```

```python
response_3=with_message_history.invoke([
    HumanMessage(
        content="Hi, my name is Isaak."
    )
],
    config=config_3
)
```

```python
response_3.content
```

### Integrate Prompt Template with Chat History

```python
prompt=ChatPromptTemplate.from_messages([
    ("system","You are a helpful assistant. Answer all the questions to the best of your ability in {language}."),
    MessagesPlaceholder(variable_name='messages')
])
```

```python
chain=prompt|llm
```

```python
response_4 = chain.invoke({
    "messages":[
        HumanMessage(
            content='Hi, my name is Isaak.'
        )
    ],
    "language":"German"
})
```

```python
response_4.content
```

### Integrate Input Messages Keys

```python
with_message_history=RunnableWithMessageHistory(
    chain,
    get_session_history,
    input_messages_key="messages"
)
```

```python
config_4={'configurable':{'session_id':'chat4'}}
```

```python
response_5 = with_message_history.invoke({
    "messages":[
        HumanMessage(
            content='Hi, my name is Isaak.'
        )
    ],
    "language":"German"
},
    config=config_4
)
```

```python
response_5.content
```

```python
response_6 = with_message_history.invoke({
    "messages":[
        HumanMessage(
            content='Hi, my name is Isaak.'
        )
    ],
    "language":"Swedish"
},
    config=config_4
)
```

```python
response_6.content
```
